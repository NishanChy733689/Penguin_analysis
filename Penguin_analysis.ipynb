```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
```

```python
df
```

```python
df.isnull().values.any()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.isnull().values.any()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
sns.pairplot(df,hue='island')
```

```python
sns.pairplot(df,hue='species')
```

```python
sns.scatterplot(data=df,x='bill_depth_mm',y='body_mass_g')
```

```python
sns.scatterplot(data=df,x='body_mass_g',y='flipper_length_mm')
```

```python
plt.figure(figsize=(20,6))
sns.jointplot(data=df,x='body_mass_g',y='flipper_length_mm',kind='reg',ci=100)

plt.show()
```

```python
sns.boxplot(x=df.species,y=df.body_mass_g,hue=df.sex,palette='rocket_r')
```

```python
plt.figure(figsize=(20,6))
sns.jointplot(data=df,x='body_mass_g',y='flipper_length_mm',kind='hex')

plt.show()
```

```python
plt.figure(figsize=(1|8,9))
sns.kdeplot(data=df,x='body_mass_g',y='flipper_length_mm',kind='reg')

plt.show()
```

```python
sns.swarmplot(x=df.species,y=df.flipper_length_mm)
```

```python
sns.rugplot(df['body_mass_g'])
```

```python
sns.countplot(x=df.island)
```

```python
v=sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
v.plot_joint(sns.kdeplot,
             fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=100, cmap="YlGnBu")
```
