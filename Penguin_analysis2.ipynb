```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://drive.google.com/uc?export=download&id=1ThuAjMaUDKWQV6y36OZ4uHYBKtbmt1xs')
```

```python

df = data.drop([ 'Stage', 'Individual ID', 'Clutch Completion', 'Date Egg', 'Comments','Delta 15 N (o/oo)','Delta 13 C (o/oo)'], axis=1)
```

```python
df.columns
```

```python
df.info()
```

```python
vals=[]
restricted=['.','nan']
for i in df['Gender']:
    vals.append(i)
vals
x=[]
for i in vals:
  if i not in restricted:
   if i == 'MALE':
    x.append('Male')
   elif i == 'FEMALE':
    x.append('Female')
   else:
    x.append(i)
x
```

```python
sns.countplot(x='Island', data=df)
plt.show()
plt.figure(figsize=(15,6))
sns.countplot(x='Species', data=df)
plt.show()
sns.countplot(x=x)
```

```python
sns.barplot(x="studyName", y="Body Mass (g)", data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='Species')
sns.pairplot(df,hue='Island')
```
